```python
import pandas as pd
```

```python
# df variable e csv er data gula joma rakhsi
df = pd.read_csv('dataset.csv')

df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#input age, gender
x = df.drop(columns=['genre'])

# output genre
y = df['genre']

# model nam disi oke
model = DecisionTreeClassifier()

#training
model.fit(x.values, y)
```

```python
# testing
# age 20 gender 1

prediction = model.predict([ [75,0]])
prediction
```

```python
# trainig & testing and performance of our model

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model.fit(x_train,y_train)


predictions = model.predict(x_test)

score = accuracy_score(y_test, predictions)
score
```

```python
# persisting the model because training need huge time due to large
#dataset so we save this model to a file

import joblib

# stores our model
joblib.dump(model, 'music-recommender.joblib')
```
